```python
def calcfactor(data,x0):
    # Extrahiere die Positionswerte der Peaks
    positions = data[:, 0]
    positions = np.sort(positions)

    # Berechne die Abstände zwischen den Peaks
    distances = np.diff(positions)

    # Berechne den mittleren Abstand
    mean_distance = np.mean(distances)

    # Berechne die Standardabweichung des Abstands
    std_distance = np.std(distances)

    #Berechne den Kalibrierungsfaktor
    scale2 = 128*0.246/mean_distance
    print(scale2)

    #Für die Umrechnung in Gwyddeon
    #scalegwyd = 0.246/mean_distance
   
    #Berechne Winkel des Höhenprofils
    m = (x0[2]-x0[3])/(x0[1]-x0[0])
    angle = np.arctan(m)
    beta = np.degrees(angle)

    print(np.round(mean_distance,2),"&",np.round(std_distance,2),"&",np.round(beta,2),"&",np.round(scale2,2),"\\\\")
```
