```python
def calcfactor(data,x0):
    # Extrahiere die Positionswerte der Peaks
    positions = data[:, 0]

    # Berechne die Abstände zwischen den Peaks
    distances = np.diff(positions)

    # Berechne den mittleren Abstand
    mean_distance = np.mean(distances)

    # Berechne die Standardabweichung des Abstands
    std_distance = np.std(distances)

    #print("Mittlerer Abstand zwischen den Peaks:", mean_distance)
    #print("Standardabweichung des Abstands zwischen den Peaks:", std_distance)

    end = (mean_distance/2)/np.cos(np.radians(30))
    #print(end)

    bond_length = 0.14
    scale = (end/bond_length)/128

    scale_factor = scale*np.cos(np.radians(30))

    m = (x0[2]-x0[3])/(x0[1]-x0[0])
    angle = np.arctan(m)
    beta = np.abs(np.degrees(angle))
    alpha = 180-beta

    #print("alpha",alpha,"beta", beta)

    xscale = np.cos(np.radians(beta))*scale_factor
    yscale = np.sin(np.radians(beta))*scale_factor

    #print("Gesamt",scale_factor,"x-faktor",xscale,"y-faktor",yscale)

    print(np.round(mean_distance,2),"&",np.round(std_distance,2),"&",np.round(beta,2),"&",np.round(xscale,2),"&",np.round(yscale,2),"\\\\")
```
