```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
#Import 
tipcur = np.genfromtxt("A1/tipcur.csv", delimiter=",")
topo = np.genfromtxt("A1/topo.csv", delimiter=",")
```

```python
x, y, z = tipcur[:,0], tipcur[:,1],tipcur[:,2]

plt.imshow(z.reshape(len(y), 1), cmap='cividis', aspect='auto', origin='lower', extent=[x.min(), x.max(), y.min(), y.max()])

# Achsenbeschriftungen
plt.xlabel('X')
plt.ylabel('Y')

# Farbbar
plt.colorbar(label='Z')

# Titel
plt.title('2D Heatmap der Daten')

# Zeige den Plot
plt.show()
```
